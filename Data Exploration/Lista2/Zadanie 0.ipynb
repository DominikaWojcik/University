```python
import numpy as np
import scipy
```

```python
a = np.array(range(1,101), dtype=np.int)
print(a)
```

```python
b = np.array(range(1,100,2), dtype=np.int)
print(b)
```

```python
c = np.array([np.pi * f * 0.01 for f in range(-100,101)], dtype=np.float64)
print(c)
```

```python
d = np.array([np.pi * f * 0.01 for f in range(-100,101) if f != 0], dtype=np.float64)
print(d)
```

```python
e = np.array([max(np.sin(i), 0.0) for i in range(1,101)], dtype=np.float64)
print(e)
```

```python
A = np.array(range(1,101), dtype=np.int).reshape((10,10))
print(A)
```

```python
B = np.diag(range(1,101)) + np.diag(range(99,0,-1),k=1) + np.diag(range(99,0,-1),k=-1)
print(B)
```

```python
C = np.array([[1 if i <= j else 0 for j in range(1,101)] for i in range(1,101)], dtype=np.int)
print(C)
```

```python
D = np.array([range(1,101), [np.math.factorial(i) for i in range(1,101)]])
print(D)
```

```python
E = np.array([[1 if j % i == 0 else 0 for j in range(1,101)] for i in range(1,101)])
print(E)
```
