```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Z TW COSINUSOW MOZEMY WYLICZYC PROMIEN WIELOKATA FOREMNEGO O BOKU d
#
# r = sqrt(d / (2*(1 - cos(alfa))))
#
def generateRandom(K, d):
    pointsInCloud = 1000
    alpha = 2.* np.pi / K
    r = np.sqrt(d / (2*(1 - np.cos(alpha))))
    vec = np.array([[r],[0.]])
    rotMatrix = np.array([[np.cos(alpha), -np.sin(alpha)], [np.sin(alpha), np.cos(alpha)]])
    
    for i in range(K):
        a = vec[0][0]
        b = vec[1][0]
        sigma = 1.
        
        X = np.random.normal(a, sigma, pointsInCloud)
        Y = np.random.normal(b, sigma, pointsInCloud)
        color = np.random.rand(3,1)
        plt.scatter(X, Y, c=color, marker='o', lw = 0)
    
        vec = np.dot(rotMatrix, vec)
    plt.show()
```

```python
generateRandom(7,5)
```

```python
generateRandom(11,10)
```

```python
generateRandom(23,15)
```
