```python
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

iris = datasets.load_iris()
```

```python
colormap={'setosa': 'Red', 'versicolor':'Green', 'virginica':'None'}
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=[colormap[i] for i in iris.target_names[iris.target]], marker='o', lw = 0)
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xlim(3.,9.)
plt.ylim(1.,5.)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.show()
```

```python
colormap={'setosa': 'Red', 'versicolor':'Green', 'virginica':'None'}
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=[colormap[i] for i in iris.target_names[iris.target]], marker='o', lw = 0)
f = np.vectorize(lambda x : 2*x - 8)
plt.plot(np.arange(3.,9.1,0.1), f(np.arange(3.,9.1,0.1)), c='black')
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xlim(3.,9.)
plt.ylim(1.,5.)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.show()
```

```python
f = np.vectorize(lambda x : 2*x - 8)

def mapColor(i):
    x = iris.data[i,0]
    y = iris.data[i,1]
    name = iris.target_names[iris.target[i]]
    if name == 'setosa':
        return 'Red' if y < f(x) else 'Green'
    elif name == 'versicolor':
        return 'Red' if y > f(x) else 'Green'
    return 'None'
    
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=[mapColor(i) for i in range(len(iris.data))], marker='o', lw = 0)
plt.plot(np.arange(3.,9.1,0.1), f(np.arange(3.,9.1,0.1)), c='black')
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xlim(3.,9.)
plt.ylim(1.,5.)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.show()
```
