```python
import numpy as np
from sklearn import datasets

iris = datasets.load_iris()

print("IRIS.DATA")
print(iris.data)
print("IRIS.TARGET")
print(iris.target)
print("IRIS.FEATURE_NAMES")
print(iris.feature_names)
print("IRIS.TARGET_NAMES")
print(iris.target_names)
```

```python
import matplotlib.pyplot as plt

plt.plot(iris.data[:, 0], iris.data[:, 1], 'ro')
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.show()
```

```python
plt.plot(iris.data[:, 0], iris.data[:, 1], 'ro')
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xlim(3.,9.)
plt.ylim(1.,5.)
plt.show()
```

```python
plt.plot(iris.data[:, 0], iris.data[:, 1], 'ro')
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xlim(3.,9.)
plt.ylim(1.,5.)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.show()
```

```python
colormap={0: 'Red', 1:'Blue', 2:'Green'}
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=[colormap[i] for i in iris.target], marker='o', lw = 0)
plt.title("Iris")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.xlim(3.,9.)
plt.ylim(1.,5.)
plt.xticks(range(3,10))
plt.yticks(range(1,6))
plt.show()
```
