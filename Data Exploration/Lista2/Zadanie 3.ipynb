```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(1,11)
Y = np.array([10,10,11,12,18,18,19,26,19,26])

a = 2
b = 5
f = np.vectorize(lambda x: a*x + b)

plt.plot(X,Y, 'ro')
plt.plot(np.arange(0,12.1,0.1), f(np.arange(0,12.1,0.1)), 'black')
plt.show()
```

```python
# odl punktu (x_0,y_0) od prostej Ax + By + C = 0 to 
# | Ax_0 + By_0 + C |
#----------------------
#   sqrt(A^2 + B^2)
#
# dla y = ax + b mamy postać (-a)*x + 1*y - b = 0
def distF(x,y):
    return np.abs(-a*x + 1*y - b) / np.sqrt((-a)**2 + 1**2)

def printSum():    
    suma = 0.0
    for i in range(len(X)):
        suma += distF(X[i],Y[i])
    print(suma)
    
printSum()
```

```python
a = 2
b = 6
f = np.vectorize(lambda x: a*x + b)

printSum()

plt.plot(X,Y, 'ro')
plt.plot(np.arange(0,12.1,0.1), f(np.arange(0,12.1,0.1)), 'black')
plt.show()
```
